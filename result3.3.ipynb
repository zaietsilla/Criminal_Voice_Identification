```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
import numpy as np

# Оновлені значення для класифікації
true_positives = 15  # TP: зловмисник виявлений у записах, де він присутній
false_positives = 1  # FP: зловмисник виявлений у записах, де його немає
true_negatives = 40  # TN: зловмисника не виявлено у записах, де його немає
false_negatives = 0  # FN: зловмисника не виявлено у записах, де він присутній

# Створення матриці невідповідностей
confusion_mat = np.array([[true_positives, false_negatives], [false_positives, true_negatives]])

# Візуалізація матриці невідповідностей
plt.figure(figsize=(8, 6))
sns.heatmap(confusion_mat, annot=True, fmt="d", cmap='Blues',
            xticklabels=["Записи на яких є зловмисник", "Записи без зловмисників"],
            yticklabels=["", ""])
plt.ylabel('')
plt.xlabel('')
plt.title('Матриця невідповідностей')
plt.show()
```

```python
# Обчислення метрик
accuracy = (true_positives + true_negatives) / (true_positives + false_positives + true_negatives + false_negatives)
recall = true_positives / (true_positives + false_negatives)
precision = true_positives / (true_positives + false_positives)
f1_score = 2 * (precision * recall) / (precision + recall)

accuracy, recall, precision, f1_score
```

```python
import pandas as pd

df_result = pd.DataFrame([{
    'Accuracy': f"{accuracy*100:.2f}%",
    'Recall': f"{recall*100:.2f}%",
    'Precision': f"{precision*100:.2f}%",
    'F1-Score': f"{f1_score*100:.2f}%",
}])

display(df_result)
```
